### Tasks 1-2

(a) Exact derivative approach

```python
import math

f = lambda x:math.atan(x)
df = lambda x:1/(1+x*x)

def newton(x, f, df):
    return x - f(x)/df(x)

x = 1.3
for _ in range(6):
    x_old = x
    x = newton(x, f, df)
    print(x, abs(x)/abs(x_old)**3) # cubic

# Binary search to find convergence interval

l = 1
r = 2
while r-l>1e-15:
    x0 = (r+l)/2
    if abs(newton(x0, f, df)) > abs(x0):
        r = x0
    else:
        l = x0
print("Convergence interval endpoints:", -x0, "and", x0)   
```

(b) Finite differences approach

```python
h = 0.0000001
f = lambda x:math.atan(x)
df = lambda x:(f(x + h) - f(x - h))/(2*h)

def newton(x, f, df):
    return x - f(x)/df(x)

x = 1
for _ in range(7):
    x_old = x
    x = newton(x, f, df)
    print(x, abs(x)/abs(x_old)**1) # linear

# Binary search to find convergence interval

l = 1
r = 2
while r-l>1e-15:
    x0 = (r+l)/2
    if abs(newton(x0, f, df)) > abs(x0):
        r = x0
    else:
        l = x0
print("Convergence interval endpoints:", -x0, "and", x0) 
```

(c) Secant method approach

```python
f = lambda x:math.atan(x)

def newton(x, x_old, f):
    return (x - f(x)*((x-x_old)/(f(x)-f(x_old))), x)

x_old = 1
x = 4
for _ in range(7):
    x, x_old = newton(x, x_old, f)
    print(x, abs(x)/abs(x_old)**2) # quadratic
```
